```python
total_score_df=delivery.groupby(['match_id','innings']).sum()['total_runs'].reset_index()
```

```python
total_score_df=total_score_df[total_score_df['innings'] == 1]
total_score_df.shape
total_score_df
```

```python
match_df=match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
match_df
```

```python
match_df['team1'].unique()
```

```python
teams= [
    'India',
    'England', 
    'New Zealand', 
    'Australia', 
    'Sri Lanka',
    'South Africa', 
    'Bangladesh', 
    'Pakistan', 
    'Nepal', 
    'Afghanistan',
    'West Indies', 
    'Scotland', 
    'United Arab Emirates', 
    'Oman',
    'Netherlands', 
    'Zimbabwe', 
    'United States of America', 
    'Ireland',
    'Canada', 
    'Namibia',
    'Papua New Guinea', 
    'Hong Kong',
    'Kenya'
]
```

```python
    match_df = match_df[match_df['team1'].isin(teams)]
    match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df=match_df[match_df['dl_applied'] == 0]

match_df
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df=delivery_df[delivery_df['innings'] == 2]
delivery_df.shape
print()
delivery_df.head()
```

```python
delivery_df['currentscore']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
# deliverydf['currentscore'] = deliverydf.groupby('match_id')['total_runs_y'].cumsum()
delivery_df.shape
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['currentscore']
delivery_df
```

```python
delivery_df['balls_left']=300 - (delivery_df['over'] * 6 + delivery_df['ballofover'])
# deliverydf['balls_left'] = 300 - (deliverydf['over']*6 + deliverydf['ballofover'])
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets_left'] = 10 - wickets
delivery_df.tail(6)
```

```python
#crr
delivery_df['crr']=(delivery_df['currentscore'] * 6)/(300 - delivery_df['balls_left'])
```

```python
#rrr
delivery_df['rrr']=(delivery_df['runs_left'] * 6)/(delivery_df['balls_left'])
```

```python
#winner
def result(row) : 
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
final_df = final_df.sample(final_df.shape[0])
final_df
```

```python
final_df.dropna(inplace=True)
final_df=final_df[final_df['balls_left'] != 0]
```

```python
from sklearn.model_selection import train_test_split
X=final_df.iloc[:,:-1]
Y=final_df.iloc[:,-1]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
